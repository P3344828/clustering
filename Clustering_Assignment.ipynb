Q1. What is unsupervised learning in the context of machine learning?

- Ans: Unsupervised learning is a type of machine learning where algorithms analyze unlabeled data to uncover hidden patterns, structures, or relationships without predefined outputs. Given data (X = {x1,x2,‚Ä¶,xn}) with no target lables, the goal is to model the underlying distribution (p(x)) or group similar data points.
- Main types:
  - Clustering (e.g., K-Means)
  - Dimensionality Reduction (e.g., PCA)
  - Association Rule Learning (e.g., Apriori)

Q2. How does K-Means clustering algorithm work?

- Ans: K-Means clustering is an unsupervised learning algorithm that partitions a dataset into (K) clusters by minimizing the within-cluster sum of squares (WCSS). Given data (X={x1,x2,‚Ä¶,xn}), K-Means aims to find cluster centroids {Œº1,Œº2,‚Ä¶,ŒºK} that minimize:
                 
                 J = ‚àëK‚Äã xi‚Äã‚ààCj ‚Äã‚àë ‚Äã‚à£‚à£xi‚Äã‚àíŒºj‚Äã‚à£‚à£2

- Algorithm steps:
  - Intialize (K) centroids randomly.
  - Assign each data point to the nearest centroid.
  - Update each centroid as the mean of its assigned points.
  - Repeat steps 2-3 until convergence.

Q3. Explain the concept of a dendrogram in hierarchical clustering.

- Ans: A dendrogram is a tree-structured diagram that illustrates the hierarchical relationships among data points in hierarchical clustering. Formally, it represents the sequence of cluster merges (or splits) based on a distance metric and linkage criterion. Each leaf node corresponds to an individual data point, and internal nodes represent clusters formed by merging smaller ones. The height of each merge indicates the dissimilarity between the clusters.

Q4. What is the main difference between K-Means and Hierarchical Clustering?

- Ans: The main difference between K-Means and Hierarchical Clustering lies in their approach to forming clusters.
 - K-Means is a partitional method that divides data into a predefined number of ùêæ clusters by minimizing intra-cluster variance.
 - Hierarchical Clustering is a hierarchical method that builds a tree-like structure (dendrogram) by successively merging or splitting clusters based on similarity, without requiring a predefined number of clusters.

Q5. What are the advantages of DBSCAN over K-Means?

- Ans: Advantages of DBSCAN over K-Means:
  - No need to predefine cluster number: DBSCAN determines clusters based on density.
  - Detects arbitrary-shaped clusters: Not limited to spherical clusters.
  - Handles noise and outliers: Can label points as noise.
  - Independent of intialization: Results are not affected by starting points.

Q6. When would you use Silhouette Score in clustering?

- Ans: The Silhouette Score is a metric to evaluate clustering quality by measuring how well each point fits within its cluster compared to other clusters. Formally, for a point ùëñ:

                s(i) = (b(i) - a(i)) / (max(a(i),b(i)))

- where ùëé(ùëñ) is the average intra-cluster distance and ùëè(ùëñ) is the minimum average distance to other clusters. Scores range from ‚àí1 (misclassified) to 1 (well-clustered).

Q7. What are the limitations of Hierarchical Clustering?

- Ans: Limitations of Hierarchical Clustering:
  - Computationally expensive: Time and memory complexity increase rapidly with dataset size.
  - Sensitive to noise and outliers: Can distort cluster structure.
  - No reassignment of points: Once merged or split, cluster membership cannot be changed.
  - Difficulty in choosing the number of clusters: Requires cutting the dendrogram at an appropriate level, which may be arbitrary.

Q8. Why is feature scaling important in clustering algorithms like K-Means?

- Ans: Feature scaling is important in clustering algorithms like K-Means because the algorithm uses distance measures (e.g., Euclidean distance) to assign points to clusters.
- If features have different scales, those with larger ranges will dominate the distance calculation, leading to biased or distorted clusters. Scaling ensures that all features contribute equally, resulting in more accurate and balanced clustering.

Q9. How does DBSCAN identify noise points?

- Ans: DBSCAN identifies noise points as those data points that do not belong to any cluster, meaning they have fewer than 'minPts' neighbors within the `Œµ`
(epsilon) radius.
- A point (p) is noise if (‚à£NŒµ‚Äã(p)‚à£< minPts) and it is not reachable from any core point, where NŒµ‚Äã(p) is the Œµ-neighborhood of (p).

Q10. Define inertia in the context of K-Means.

- Ans: In the context of K-Means clustering, inertia is a measure of how internally coherent the clusters are, defined as the sum of squared distances between each point and the centroid of its assigned cluster. For a dataset,
D ={x1‚Äã,x2‚Äã,‚Ä¶,xn‚Äã} and cluster centroids {Œº1,Œº2,‚Ä¶,Œºk}:

                Inertia = ‚àëk‚Äãx‚ààCi ‚Äã‚àë‚Äã‚à•x‚àíŒºi‚Äã‚à•2

- Where ùê∂ùëñ is the set of points in cluster ùëñ, and ‚à•‚ãÖ‚à• denotes the Euclidean distance.
- Lower inertia indicates tighter, more compact clusters, while higher inertia indicates more spread-out clusters.

Q11. What is the elbow method in K-Means clustering?

- Ans: The Elbow Method is a heuristic used in K-Means clustering to determine the optimal number of clusters (k). It involves plotting the Within-Cluster Sum of Squares (WCSS) against different values of (k) and identifying the point where the rate of decrease sharply changes, forming an elbow. This point indicates the optimal (k), beyond which adding more clusters yields marginal improvement in clustering performance.

Q12. Describe the concept of "density" in DBSCAN.

- Ans: In DBSCAN, density refers to the concentration of data points within a region. A region is considered dense if it contains at least a minimum number of points (MinPts) within a specified radius (Œµ). Clusters are formed from these dense regions, while points in low-density areas are treated as noise.

Q13. Can hierarchical clustering be used on categorical data?

- Ans: Yes, hierarchical clustering can be applied to categorical data provided that an appropriate similarity or distance measure is used, such as the Jaccard coefficient, Hamming distance, or simple matching coefficient. Since traditional hierarchical clustering relies on numerical distance metrics, categorical variables must first be represented or encoded in a form compatible with these measures.

Q14. What does a negative Silhouette Score indicate?

- Ans: A negative Silhouette Score ndicates that, on average, data points are closer to neighboring clusters than to their own assigned cluster. This suggests poor clustering quality, where many points may be misclassified or clusters are overlapping and not well separated.

Q15. Explain the term "linkage criteria" in hierarchical clustering.

- Ans: In hierarchical clustering, the linkage criterion defines the method used to calculate the distance between clusters during the merging process. It determines how individual point-to-point distances are aggregated to measure inter-cluster dissimilarity. Common linkage criteria include single linkage (minimum distance), complete linkage (maximum distance), average linkage (mean distance), and Ward's linkage (minimization of within-cluster variance).

Q16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?

- Ans: K-Means clustering may perform poorly on datasets with clusters of varying sizes or densities because it assumes clusters are spherical, similarly sized, and equally dense. When these assumptions are violated, K-Means can misclassify points, merge smaller or sparser clusters into larger ones, and fail to accurately capture the true cluster structure.

Q17. What are the core parameters in DBSCAN, and how do they influence clustering?

- Ans: The core parameters in DBSCAN are:
  -  Œµ (epsilon): The neighborhood radius that defines how close points must be to each other to be considered part of the same cluster.
    - Influence: Smaller Œµ ‚Üí tighter clusters, more noise points; larger Œµ ‚Üí larger clusters, fewer noise points.
  - min_samples: The minimum number of points required to form a dense region (cluster).
    - influence: Smaller min_samples ‚Üí more clusters, fewer noise points; larger min_samples ‚Üí fewer clusters, more noise points.

Q18. How does K-Means++ improve upon standard K-Means intialization?

- Ans: K-Means++ improves standard K-Means by intializing centroids more strategically: the first centroid is chosen randomly, and each subsequent centroid is selected with probability proportional to its squared distance from the nearest existing centroid. This ensures well-spread intial centroids, leading to faster convergence and higher-quality clustering.

Q19. What is agglomerative clustering?

- Ans: Agglomerative clustering is a hierarchical clustering technique that constructs a hierarchy of clusters in a bottom-up manner, starting with each data point as an individual cluster and iteratively merging the closest clusters based on a specified linkage criterion and distance metric, until all points are grouped or a predefined stopping condition is met.

Q20. What makes Silhouette Score a better metric than just inertia for model evaluation?

- Ans: Silhouette Score is superior to inertia because it evaluates both intra-cluster cohension and inter-cluster separation, whereas inertia measures only cluster compactness and decreases with more clusters. This makes Silhouette Score a more robust and comparable metric for assessing clustering quality.

```python
# Q21. Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot.

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Generate synthetic data with 4 centers
x, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=1)

# Apply K-Means clustering & Visualize Clusters
kmeans = KMeans(n_clusters=4, random_state=1)
y_kmeans = kmeans.fit_predict(x)
plt.scatter(x[:, 0], x[:, 1], c=y_kmeans, s=50, cmap='viridis')

# Plot centroids
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='X', label='Centroids')
plt.title('K-Means Clustering on Synthetic Data (4 Centers)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
# Q24. Load the Wine dataset and apply K-Means clustering after standardizing the features. Print the size of each cluster.

from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import numpy as np

# Load the Wine dataset & Standardize the features
wine = load_wine()
X = wine.data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply K-Means clustering & the size of each cluster
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X_scaled)
unique, counts = np.unique(labels, return_counts=True)
print("Cluster sizes:")
for cluster, size in zip(unique, counts):
    print(f"Cluster {cluster}: {size} samples")
```

```python
# Q29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Generate synthetic data & Evaluate silhouette scores for k = 2 to 5
x, y_true = make_blobs(n_samples=500, centers=4, cluster_std=0.6, random_state=1)
scores = []
k_values = range(2, 6)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=1)
    labels = kmeans.fit_predict(x)
    score = silhouette_score(x, labels)
    scores.append(score)
    print(f"k = {k}, Silhouette Score = {score:.3f}")

# Silhouette scores as a bar chart
plt.figure(figsize=(8, 5))
plt.bar(k_values, scores, color='blue', edgecolor='black')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores for Different k values')
plt.xticks(k_values)
plt.show()
```

```python
# Q31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries.

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic data with overlapping clusters & Apply K-Means clustering
X, y_true = make_blobs(n_samples=500, centers=3, cluster_std=2.0, random_state=1)
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)
centers = kmeans.cluster_centers_

# Visualize clusters with decision boundaries (create meshgrid)
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.05),
                     np.arange(y_min, y_max, 0.05))

# Predict cluster for each point in meshgrid & Plot decision boundaries
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap='viridis')
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis', edgecolor='k')
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='X', label='Centroids')
plt.title('K-Means Clustering with Overlapping Clusters')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
# Q34. Load the Breast Cancer dataset and compare inertia values for k = 2 to 6 using K-Means. Show results in a line plot.

from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load and scale the data  & Compute inertia for k = 2 to 6
data = load_breast_cancer()
X = data.data
X_scaled = StandardScaler().fit_transform(X)
inertia_values = []
k_values = range(2, 7)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=1)
    kmeans.fit(X_scaled)
    inertia_values.append(kmeans.inertia_)
    print(f"k={k}, inertia={kmeans.inertia_:.2f}")

# Plot inertia vs k
plt.figure(figsize=(8, 5))
plt.plot(k_values, inertia_values, marker='o', linestyle='-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('K-Means Inertia for Different k Values')
plt.xticks(k_values)
plt.show()
```

```python
# Q40. Load the Digits dataset, apply PCA to reduce to 3 components, then use KMeans,and visualize with a 3D scatter plot.

from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Need for 3D plot

# Load Digits data & Reduce dimensionality to 3D using PCA
digits = load_digits()
X = digits.data
pca = PCA(n_components=3, random_state=42)
X_pca = pca.fit_transform(X)

# Apply K-Means clustering & 3D scatter plot
kmeans = KMeans(n_clusters=10, random_state=1)
labels = kmeans.fit_predict(X_pca)
centers = kmeans.cluster_centers_

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(X_pca[:,0], X_pca[:,1], X_pca[:,2], c=labels, cmap='tab10', s=50, alpha=0.6)
ax.scatter(centers[:,0], centers[:,1], centers[:,2], c='red', s=200, marker='X', label='Centroids')
ax.set_title('K-Means Clustering on Digits (3D PCA)')
ax.set_xlabel('PCA 1')
ax.set_ylabel('PCA 2')
ax.set_zlabel('PCA 3')
ax.legend()
plt.show()
```

```python
# Q42. Load the Breast Cancer dataset, reduce dimensionality using PCA, and apply Agglomerative Clustering. Visualize in 2D.

from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

# Load and scale data & Reduce dimensionality to 2D using PCA
data = load_breast_cancer()
X = data.data
X_scaled = StandardScaler().fit_transform(X)
pca = PCA(n_components=2, random_state=1)
X_pca = pca.fit_transform(X_scaled)

# Apply Agglomerative Clustering & Visualize
agg = AgglomerativeClustering(n_clusters=2, linkage='ward')
labels = agg.fit_predict(X_pca)
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0], X_pca[:,1], c=labels, cmap='viridis', s=50)
plt.title('Agglomerative Clustering on Breast Cancer Dataset (PCA 2D)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()
```

```python
# Q43. Generate noisy circular data using make_circles and visualize clustering results from KMeans and DBSCAN side-by-side.

from sklearn.datasets import make_circles
from sklearn.cluster import KMeans, DBSCAN
import matplotlib.pyplot as plt

# Generate noisy circular data & Apply K-Means
X, y_true = make_circles(n_samples=500, factor=0.5, noise=0.05, random_state=1)
kmeans = KMeans(n_clusters=2, random_state=1)
labels_kmeans = kmeans.fit_predict(X)

# Apply DBSCAN & Plot side-by-side & K-Means plot & DBSCAN plot
dbscan = DBSCAN(eps=0.1, min_samples=5)
labels_dbscan = dbscan.fit_predict(X)

fig, axs = plt.subplots(1, 2, figsize=(14,6))
axs[0].scatter(X[:,0], X[:,1], c=labels_kmeans, cmap='viridis', s=50)
axs[0].set_title('K-Means Clustering')
axs[0].set_xlabel('Feature 1')
axs[0].set_ylabel('Feature 2')

axs[1].scatter(X[:,0], X[:,1], c=labels_dbscan, cmap='viridis', s=50)
axs[1].set_title('DBSCAN Clustering')
axs[1].set_xlabel('Feature 1')
axs[1].set_ylabel('Feature 2')

plt.show()
```

```python
# Q44. Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering.

from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples
import matplotlib.pyplot as plt
import numpy as np

# Load Iris data & Apply K-Means clustering
iris = load_iris()
X = iris.data
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)

# Compute silhouette scores & plot silhouette coefficients
silhouette_vals = silhouette_samples(X, labels)
avg_score = silhouette_vals.mean()
print(f"Average Silhouette Score: {avg_score:.3f}")

plt.figure(figsize=(10,6))
y_lower = 10

for i in range(3):
    ith_cluster_silhouette = silhouette_vals[labels == i]
    ith_cluster_silhouette.sort()
    size_cluster = ith_cluster_silhouette.shape[0]
    y_upper = y_lower + size_cluster

    plt.fill_betweenx(np.arange(y_lower, y_upper),
                      0, ith_cluster_silhouette,
                      alpha=0.7)
    plt.text(-0.05, y_lower + 0.5 * size_cluster, str(i))
    y_lower = y_upper + 10
plt.axvline(x=avg_score, color="red", linestyle="--", label="Average Silhouette")
plt.xlabel("Silhouette Coefficient")
plt.ylabel("Sample Index")
plt.title("Silhouette Plot for K-Means Clustering on Iris Dataset")
plt.legend()
plt.show()
```

```python
# Q46. Load the Wine dataset, apply KMeans, and visualize the cluster assignments in a seaborn pairplot (first 4 features).

from sklearn.datasets import load_wine
from sklearn.cluster import KMeans
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load Wine dataset & Convert to dataframe for seaborn
wine = load_wine()
x = wine.data[:, :4]
feature_names = wine.feature_names[:4]
df = pd.DataFrame(x, columns=feature_names)

# Apply K-Means clustering & Add cluster labels to dataframe
kmeans = KMeans(n_clusters=3, random_state=1)
labels = kmeans.fit_predict(x)
df['Cluster'] = labels.astype(str)

# Pairplot colored by cluster
sns.pairplot(df, hue='Cluster', palette='viridis', diag_kind='kde')
plt.suptitle('K-Means Clustering on Wine Dataset (First 4 Features)', y=1.02)
plt.show()
```

```python
# Q47. Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the count.

from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
import numpy as np

# Generate noisy blobs & Apply DBSCAN
X, y_true = make_blobs(n_samples=500, centers=4, cluster_std=1.2, random_state=1)
dbscan = DBSCAN(eps=0.8, min_samples=10)
labels = dbscan.fit_predict(X)

# Count clusters (excluding noise) and noise points
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
n_noise = np.sum(labels == -1)

print(f"Number of clusters (excluding noise): {n_clusters}")
print(f"Number of noise points: {n_noise}")
```
